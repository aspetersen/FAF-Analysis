```python
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
melted_data = pd.melt(dfFAFAgg, value_vars=tons, var_name='Year', value_name='Tons')
dfAggTonsfiltered = melted_data.groupby(['Year'],as_index=False)['Tons'].sum()
dfAggTonsfiltered

```

```python

# Create dropdown widgets for filtering
mdlcom_des_dropdown = widgets.Dropdown(options=['All', 'AGRI', 'BULD', 'CHEM', 'COAL', 'FOOD', 'LRET', 'MANU', 'MNRL', 'OLGA', 'PETR', 'TEXT'], value='All', description='Model Description: ')
mode_desc_dropdown = widgets.Dropdown(options=['All', 'Truck', 'Multiple modes & mail', 'Rail', 'Other and unknown', 'Air (include truck-air)', 'Pipeline', 'Water', 'No domestic mode'], value='All', description='Mode: ')

# Create a function to update the plot based on dropdown selections
def update_plot(commodity, mode):
    filtered_FAF = dfFAFAgg.copy()
    
    if commodity != 'All':
        
        filtered_FAF = filtered_FAF[filtered_FAF['mdlcom_des'] == commodity]
    
    if mode != 'All':
        filtered_FAF = filtered_FAF[filtered_FAF['mode_desc'] == mode]

    # Reshape the data using melt
    melted_data = pd.melt(filtered_FAF, value_vars=tons, var_name='Year', value_name='Tons')
    dfAggTonsfiltered = melted_data.groupby(['Year'],as_index=False)['Tons'].sum()
    # Create a line plot of values by specific columns
    sns.lineplot(data=dfAggTonsfiltered, x='Year', y='Tons')
    plt.xticks(rotation=60)
    plt.title("Tons by Year")
    plt.ticklabel_format(style='plain', axis='y')


# Create an interactive plot
interactive_plot = widgets.interactive(update_plot, commodity=mdlcom_des_dropdown, mode=mode_desc_dropdown)

# Display the widgets and plot
##display(widgets.HBox([mdlcom_des_dropdown, mode_desc_dropdown]))
display(interactive_plot)

```

```python
#Filter by Utah then add column for direction(II,XI,XX,IX)
utahdmscodes = [499,491]
dfFAFAggUtah = dfFAFAgg[dfFAFAgg['dms_orig'].isin(utahdmscodes) | dfFAFAgg['dms_dest'].isin(utahdmscodes)] 

dfFAFAggUtah
```

```python
# Define a function to apply the logic and populate the "direction" column
def populate_direction(row):
    if row['dms_orig'] in [491, 499] and row['dms_dest'] in [491, 499]:
        return 'II'
    elif row['dms_orig'] in [491, 499] and row['dms_dest'] not in [491, 499]:
        return 'IX'
    elif row['dms_orig'] not in [491, 499] and row['dms_dest'] in [491, 499]:
        return 'XI'
    else:
        return 'XX'
# Apply the function to create a new column "direction"
dfFAFAggUtah['Direction'] = dfFAFAggUtah.apply(lambda row: populate_direction(row), axis=1)


dfFAFAggUtah
```
